### 7. Estimate the total revenue for a customer value using the following table 

```python
def sort_df_with_time(dataframe):
    sorted_df = dataframe[['customer_no', 'location', 'time', 'date', 'day_of_week', 'timestamp']].sort_values(by=['date', 'customer_no', 'time'])
    return sorted_df
```

```python
def time_spent(dataframe):
    dataframe['next_time'] = dataframe['timestamp'].shift(-1)
    time_spent = []
    for i, row in dataframe.iterrows():
        if row['location'] == 'checkout':
            pass
        if row['next_time'] == None:
            time_spent.append([row['customer_no'],row['location'],0])
        else:
            duration = row['next_time'] - row['timestamp']
            time_spent.append([row['customer_no'],row['location'],duration ])
    return time_spent
```

```python
monday = days[days['day_of_week'] == 0]
tuesday = days[days['day_of_week'] == 1]
wednesday = days[days['day_of_week'] == 2]
thursday = days[days['day_of_week'] == 3]
friday = days[days['day_of_week'] == 4]
```

```python
#Do this by variable df name in a for-loop
```

```python
def create_clean_timedf(dataframe):
    dataframe = pd.DataFrame(time_spent(sort_df_with_time(dataframe)),columns = ['customer_no', 'location', 'duration'])
    dataframe = dataframe[dataframe['location']!= 'checkout']
    dataframe = dataframe[dataframe['duration'] > pd.Timedelta(-82800)]
    dataframe['duration_min'] = dataframe['duration']/np.timedelta64(1, 'm')
    return dataframe
```

```python
mo_time = create_clean_timedf(monday)
tue_time = create_clean_timedf(tuesday)
wed_time = create_clean_timedf(wednesday)
thu_time = create_clean_timedf(thursday)
fri_time = create_clean_timedf(friday)
```

## ON AVERAGE: How many people enter per day per unique min

```python
entering_for_each_cust = days.groupby(['date','customer_no'])['timestamp'].min().unstack(0)

new_cust = pd.DataFrame(columns=['hour'],index=pd.date_range("2019-09-06 07:00:00", "2019-09-06 21:50:00", freq="1min"))
new_cust['hour'] = new_cust.index.time
new_cust.reset_index(drop=True, inplace=True)

for i in range(entering_for_each_cust.shape[1]):
    temp = pd.DataFrame(entering_for_each_cust.iloc[:, i].value_counts().sort_index())
    temp['hour'] = temp.index.time
    new_cust = pd.merge(new_cust, temp, how='left', left_on='hour', right_on='hour')
new_cust = new_cust.fillna(0)
new_cust['mean'] = new_cust.mean(axis = 1)
new_cust['stdev'] = new_cust.std(axis = 1)
new_cust['mean'] = new_cust['mean'].round().astype(int)
new_cust['stdev'] = new_cust['stdev'].round().astype(int)
```
