## Decision on What Type of Scaling to Do: ##
- Since I inferred meanings from all variables on mean, (except hour and year), outliers are accounted for
- Moreover, I do not want to give up on the underlying distribution
- Chosen: MinMaxScaler(

```python
Xtrain_nonscaled = df_train[['Hour' ,'Year', 'atemp']]
ytrain_nonscaled = df_train['count']
ytrain_nonscaled
```

```python
#Scale the Numerical Variables 
sc = MinMaxScaler()
Xtrain_sc= sc.fit_transform(Xtrain_nonscaled)
Xtrain_sc = pd.DataFrame(Xtrain_sc, columns = Xtrain_nonscaled.columns)
```

```python
#Take log1p of the target var
ytrain_sc = np.log1p(ytrain_nonscaled)
```

```python
ytrain_sc = pd.DataFrame(ytrain_sc)
```

```python
#Ridge
ridge = Ridge()
ridge_param = { 'max_iter':[3000],'alpha':[0.1, 1, 2, 3, 4, 10, 30,100,200,300,400,800,900,1000]}

grid_ridge_m = GridSearchCV( ridge,ridge_param,scoring = 'neg_mean_squared_error',cv=5)
grid_ridge_m.fit( Xtrain_sc, ytrain_sc )
ypred = grid_ridge_m.predict(Xtrain_sc)
print (grid_ridge_m.best_params_)


best_score = (grid_ridge_m.best_score_)*-1
print(best_score)
```

```python
#Get Coefficients
#Ridge
ridge = Ridge(alpha = 10, max_iter= 3000)
ridge.fit(Xtrain_sc, ytrain_sc)
ridge.predict(Xtrain_sc)
print(ridge.coef_)
print(Xtrain_sc.columns)
```

```python
#Check the assumptions
#Mean residual is around zero
residuals = ypred - ytrain_sc
print(residuals.mean())
residuals.hist()
```

```python
#Residuals uncorrelated with x and must be equally on and under 0

for column in Xtrain_sc.columns:
    plt.scatter(Xtrain_sc[column], residuals)
    plt.title(f'{column}')
    plt.show()
```
