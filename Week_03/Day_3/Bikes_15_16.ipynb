# Get the data with API # 

```python
import requests
from config import DARK_SKY_SECRET_KEY
from datetime import datetime
```

```python
# create a range of dates to request the data from the darksky api
date_range = pd.date_range('2015/1/1', '2016/12/31', freq='D').tz_localize('US/Eastern')
```

```python
len(date_range)
```

```python
len(new_15_16)/24 # over 24 does not work. When merging I have to do, inner merge on the datetime index!!!!
```

# weather = pd.DataFrame()

```python
for dates in date_range:
    result = requests.get(f'https://api.darksky.net/forecast/{DARK_SKY_SECRET_KEY}/38.907,-77.0417,{int(dates.timestamp())}?units=ca')
    result_json = result.json()
    df_weather = pd.DataFrame(columns=['time', 'temperature', 'atemperatue', 'humidity', 'windSpeed'])

    for entry in result_json['hourly']['data']:
        df_weather.loc[len(df_weather)] = [pd.to_datetime(entry['time'], unit='s').tz_localize('utc').tz_convert('US/Eastern'), entry['temperature'], entry['apparentTemperature'], entry['humidity'], entry['windSpeed']]
    weather = pd.concat([weather, df_weather], ignore_index = True, sort=True)
```

```python
weather
```

```python
weather.drop_duplicates(inplace = True)
```

```python
weather
```

```python
weather.head()
```

```python
merged_dataframe = pd.merge(new_15_16, weather, how = 'inner', left_index=True, right_index=True)
```

```python
merged_dataframe = merged_dataframe.loc[~merged_dataframe.index.duplicated(keep='first')]
```

```python
merged_dataframe
```

```python
merged_dataframe['Day_Name'] = merged_dataframe.index.day_name()
```

```python
merged_dataframe.head()
```

```python
#Weekday Reconvert
day_new = {'Sunday': 0, 'Tuesday': 1, 'Wednesday': 1, 'Monday': 1, 'Friday': 1, 'Thursday': 1, 'Saturday': 0}
merged_dataframe['weekday'] = merged_dataframe['Day_Name'].map(day_new)
```

```python
#Holiday Reconvert
holiday = {'Sunday': 1, 'Tuesday': 0, 'Wednesday': 0, 'Monday': 0, 'Friday': 0, 'Thursday': 0, 'Saturday': 1}
merged_dataframe['Holiday'] = merged_dataframe['Day_Name'].map(holiday)
```
