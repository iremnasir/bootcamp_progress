```python
artist = input("Enter your artist: ") 
artist_name = str(artist).lower()
artist_name_re = re.sub(' ','-',artist_name)
```

```python
df_ = []
sn = []
path = loc+artist_name_re
files = [f for f in os.listdir(path) if os.path.isfile(f)]
for f in files:
    with open(f,'r') as myfile:
        data.append(myfile.read())
        sn.append(str(f).strip())
        obj = zip(sn, data)
    df_ = pd.DataFrame(obj, columns = ['Song_name', f'{artist_name_re}'])
    df_.to_csv('metallica.csv')

```
