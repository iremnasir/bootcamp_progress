# Animate a Scatterplot
### In this exercise we will reproduce Hans Roslings famous animation. We will proceed in four steps:

 - Load the data to multiple DataFrames
 - Merge the DataFrames into one
 - Generate one scatterplot for each year.
 - Connect the images to an animated GIF.

### Life Expectency & Fertility

```python

multi_plot = df4.loc[1960:2015,[('Iran',       'Fertility'),
            ('Iran', 'Life_expectency'),('Germany',       'Fertility'),
            ('Germany', 'Life_expectency'),('Spain',       'Fertility'),
            ('Spain', 'Life_expectency')]]
multi_plot.plot(figsize = (12,8))
plt.xlabel("Year" )

plt.title("Fertility and Life Expectendy")
plt.legend(loc="best")
```

## Population:

## Gif function:

This function takes the name of a country and gives a gif of the population, fertility and life expectendy history 

```python
def gif_func(country,fps_1):
    df_1 = df6.loc[country]
    df_1 = pd.DataFrame(df_1)
    df_1 = df_1.unstack()
    for y in sel_col:
        df_new = pd.DataFrame(df_1.loc[y]).unstack()
        image = df_new.plot.scatter((y, 'Fertility'), (y,    'Life_expectency'  ),s= df_new[(y, 'Population'     )]/100000)
        plt.title(f"{country}_{y}")
        plt.ylim(0,100)
        plt.xlim(0,10)
        plt.savefig(f"iran_plots/{country}_{y}.png")
        
        list_of_images = []
    for y in sel_col:
        im = imageio.imread(f"iran_plots/{country}_{y}.png")
        list_of_images.append(im)
    imageio.mimsave(f"{country}.gif", list_of_images, fps = fps_1)    
```

```python
gif_func("Germany",10)
```
