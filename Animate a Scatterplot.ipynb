```python
import imageio


sel_col = df6.columns
sel_col = sel_col[160:]
for y in sel_col:
    df_new = df6[y].unstack(1)
    image = df_new.plot.scatter('Fertility', 'Life_expectency', s= df_new["Population"]/1000000, c = cmap)
    plt.title(y)
    plt.ylim(0,100)
    plt.xlim(0,10)
    plt.savefig(f"plots/{y}.png")
       
    
    
    

    
    
    
    
    
```

```python
list_of_images = []
for y in sel_col:
    im = imageio.imread(f"plots/{y}.png")
    list_of_images.append(im)
imageio.mimsave("animated_plot.gif", list_of_images, fps = 2)    
```
