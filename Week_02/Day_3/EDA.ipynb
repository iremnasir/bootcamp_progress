```python
#For numerical columns use mean
#For cat-cat use counts

def pl_cat(interest, variable):
    df_dummy = df_p.reset_index()
    gr = df_dummy.groupby([f'{interest}', f'{variable}']).count()
    br = gr.unstack(1)
    br.plot.bar(y = 'id')
```
