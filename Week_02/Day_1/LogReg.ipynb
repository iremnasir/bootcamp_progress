```python
from sklearn.datasets import make_moons
from sklearn.linear_model import LogisticRegression
```

```python
X, y = make_moons(n_samples=200, noise=0.3)
```

```python
X.shape, y.shape
```

```python
m = LogisticRegression(C=1e5, solver = 'lbfgs') # C is a hyperparameter (regularization, W03)
m.fit(X, y) # find the best model parameters for this data
```

```python
print(m.score(X,y)) # calculates the accuracy (% of correct points)
```

```python
m.coef_ #coeff of feature 1 and 2
```

```python
m.intercept_
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.scatter(X[:,0], X[:,1], c =y)
```

```python
point = [[2.0, -1.0]]
m.predict(point)
```

```python
m.predict_proba(point).round(4)
```
