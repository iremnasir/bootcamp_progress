#### 3. Long vs. Wide Format #### 

* long format: select single data points, move columns around
* wide format: select row/column ranges

```python
del df['initial']
```

```python
df.head()
```

```python
s = df.stack()
```

```python
s # ---> Long format
```

```python
s['Germany', 2000]
```

```python
subset = s.sample(1000) # good for randomizing every variable
subset = s.sample(1000, random_state = 42) # picking the "way" of random - aka seeding
```

```python
subset
```

- There is a way to unstack on different levels

```python


subset.unstack() # long format and converts it into a wide format (fewer rows but many columns)
```

```python
subset.unstack(0) # unstacking at different levels
```

- You can create multiple index with set_index([])

```python
data = [
    ['Germany', 2000, 80, 1.5],
    ['Germany', 2010, 81, 1.4],
    ['Germany', 2020, 82, 1.3],
    
    ['Iran', 2000, 61, 2.5],
    ['Iran', 2010, 70, 3.5],
    ['Iran', 2020, 80, 4.5]
        
        
    ]
```

```python
exercise = pd.DataFrame(data, columns = ['Country', 'year', 'pop', 'fert'])
```

```python
exercise
```

```python
exercise.set_index(['Country', 'year'], inplace = True)
```

```python
y = exercise.stack()
```

```python
y
```

```python
y.unstack(0).plot.bar()
```

```python
y.unstack(1).plot.bar()
```

```python
y.unstack(2).loc['Iran'].plot.bar()
```

```python
y.unstack((0,1)).unstack().unstack().unstack()
```
