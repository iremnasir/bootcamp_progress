```python
year_array = np.arange(1960,2016,1)
```

```python
import matplotlib.pyplot as plt
```

```python
for year in year_array:
    plt.figure(figsize=(10,7))
    df6 = df3.unstack(1)
    df6 = df6[year]
    df6 = df6.unstack(1)
    cmap = plt.get_cmap('tab20', lut = len(df6)).colors
    df6.plot.scatter('Population', 'fertility', s=df6['life_expectancy'], c=cmap)
    plt.text(5,1,year, fontsize = 16)
    plt.xlabel('Pop', fontsize = '16')
    plt.ylabel('Fert', fontsize = '16')
    plt.axis(ymin = 0, ymax= 10, xmin = 0, xmax = 2e8)
    plt.savefig(f'./plots2/plot_{year}.png', bbox = 'tight')
```
