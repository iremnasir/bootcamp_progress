```python
import numpy as np
```

```python
year_array = np.arange(1960,2016,1)
```

```python
import matplotlib.pyplot as plt
```

```python
for year in year_array:
    plt.figure(figsize=(10,7))
    df6 = df3.unstack(1)
    df6 = df6[year]
    df6 = df6.unstack(1)
    cmap = plt.get_cmap('tab20', lut = len(df6)).colors
    df6.plot.scatter('fertility', 'life_expectancy', s=df6['Population']/1000000, c=cmap)
    plt.text(5,1,year, fontsize = 16)
    plt.xlabel('Fertility', fontsize = '16')
    plt.ylabel('Life expectancy in years', fontsize = '16')
    plt.axis(ymin = 0, ymax= 90, xmin = 0, xmax = 10)
    plt.savefig(f'./plots/plot_{year}.png', bbox = 'tight')
```

```python
import imageio
```

```python
list_of_images = []
for year in year_array:
    im = imageio.imread(f'./plots/plot_{year}.png')
    list_of_images.append(im)
```
