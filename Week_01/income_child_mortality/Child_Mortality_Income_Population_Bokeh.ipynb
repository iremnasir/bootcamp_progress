```python
#Make a new df for (potentially) each year
def year_df(year, odf):
    """
    Make a dataframe for each year
    Parameters:
    -------
    year: year wanted (as string)
    odf: original stacked dataframe
    
    Returns:
    ------
    Dataframe for a given year with additional columns of continent and color code for Bokeh
    """
    year_df = odf.loc[year].unstack()
    year_df['continent'] = year_df.index.map(region_dict)
    year_df['continent'] = year_df['continent'].fillna('unknown')
    year_df = year_df.reset_index()
    year_df['income'] = year_df['income'].astype('float')
    year_df['child_mortality'] = year_df['child_mortality'].astype('float')
    year_df['population'] = year_df['population'].astype('float')
    return year_df
```
