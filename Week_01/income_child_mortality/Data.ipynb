```python
import pandas as pd
import numpy as np
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
```

```python
output_notebook()
```

```python
income = pd.read_csv('income_per_person_gdppercapita_ppp_inflation_adjusted.csv', index_col=0)
```

```python
income.head()
```

```python
child_mortality = pd.read_csv('child_mortality_0_5_year_olds_more_years_version_7.csv', index_col=0)
```

```python
child_mortality.head()
```

```python
child_mortality = child_mortality.div(1000)
```

```python
child_mortality.head()
```

```python
#Stack and join
```

```python
income_stacked = pd.DataFrame(income.stack())
```

```python
cm_stacked = pd.DataFrame(child_mortality.stack())
```

```python
cm_income = income_stacked.join(cm_stacked, how='inner', lsuffix='income', rsuffix='child_mort')
```

```python
cm_income = cm_income.rename(columns={'0income':'income', '0child_mort':'child_mortality'})
```

```python
cm_income.head()
```

```python
cm_income = cm_income.stack()
```

```python
cm_income.head()
```

```python
cm_income_unstacked = cm_income.unstack(level = (0,2))
```

```python
cm_income_unstacked.tail()
```

```python
#Create the bokeh data source
```

```python
dates = np.array(cm_income_unstacked.index, dtype=np.datetime64) # convert date strings to real datetimes
```

```python
p = figure(x_axis_type="datetime",plot_width=800, plot_height=400)
p.circle(dates, cm_income_unstacked['Turkey']['child_mortality'], radius_dimen = cm_income_unstacked['Turkey']['income'])
show(p)
```
