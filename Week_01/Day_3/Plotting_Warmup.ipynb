```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

%matplotlib inline
```

### What would be your first step?

```python
df = pd.DataFrame(np.arange(-10,10.5,.5), columns = ['x'])
#np.linspace(-10, 10, 40) includes the boundary
df['y'] = df['x'].apply(np.square)
```

### Make sure that all variables have the same length

```python
df.shape
```

### Plot with pandas and matplotlib

```python
#Pandas
df.plot.line(x = 'x', y = 'y', color = 'green', xlim = (-10.5, 10.5), ylim = (-5, 105), title = 'Polynomial $y = x^2$')
plt.xlabel('X-values')
```

```python
#Matplotlib
plt.figure(figsize=(10,7))
plt.plot(df['x'], df['y'], color= 'g', linestyle='--')
plt.legend('y')
```

#### Using Lists and Arrays ####

```python
x_values = np.linspace(-10,10,41)
```

```python
y_values = []
for x in x_values:
    y = x**2
    y_values.append(y)

np.array(y_values)
```

```python
#list comprehension
y_values = [x**2 for x in x_values]
```

## Plotting with seaborn ##

```python
df = pd.read_csv('./pokemon.csv', index_col = 0)
df.head()
```

```python
import seaborn as sns
```

```python
#Countplot Type 2
plt.figure(figsize = (10,10))
sns.countplot(x = 'Type 2', data = df, palette = 'colorblind')
plt.xticks(rotation = 90, fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel('Type 2', fontsize=16)
plt.ylabel('Count', fontsize=16)
```

```python
#Count Plot Generation
sns.set_style("ticks")
plt.figure(figsize = (10,10))
sns.countplot(x = 'Generation', data = df, palette = 'colorblind')
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel('Generation', fontsize=16)
plt.ylabel('Count', fontsize=16)
```
