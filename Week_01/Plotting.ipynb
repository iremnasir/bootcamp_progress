## Plotting with seaborn

```python
df = pd.read_csv('../descriptive_statistics/pokemon.csv', index_col=0)
df.head()
```

```python
#! pip install seaborn
```

```python
import seaborn as sns
```

### Distplot

```python
plt.figure(figsize=(12, 8))
sns.distplot(df['Attack'])
```

### Violinplot

```python
plt.figure(figsize=(12,8))

sns.violinplot(x=df.Generation,y=df.Defense,hue=df.Legendary,split=True)

plt.title("Stats")
```

###  Countplot

```python
plt.figure(figsize=(12, 8))
sns.countplot(x="Type 1", data=df,palette="rainbow")
plt.xticks(rotation=-45)
```

### KDE Plot

```python
plt.figure(figsize=(12, 8))
sns.kdeplot(df.Attack,df.Defense,shade=False,kde=True)
sns.scatterplot(df['Attack'],df['Defense'], size=0.03, legend=False, color='y')
```

### Heatmap

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True)
```

### Pairplot

```python
data = df[['Attack', 'Defense', 'Speed', 'HP', 'Legendary']]
sns.pairplot(data, hue='Legendary', diag_kind="kde", kind="scatter", palette="husl")
```
