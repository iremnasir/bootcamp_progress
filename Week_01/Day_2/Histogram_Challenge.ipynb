```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('gapminder_lifeexpectancy.xlsx', index_col = 0)
```

```python
df.head()
```

```python
lifeexp_years = df[[1950, 1975, 2000, 2015]]
lifeexp_years
```

```python
lifeexp_years[2015].hist()
plt.savefig('histo.png')
```

```python
lifeexp_years[1950].hist(bins = 5, alpha = 0.25, histtype = 'bar')
lifeexp_years[1950].hist(bins = 10)
lifeexp_years[1950].hist(bins = 20)
plt.xlabel('age')
plt.ylabel('counts')
plt.title('life expectancy in 1950')
plt.axis([25.0, 70.0, 0.0, 60.0])
plt.savefig('histogram_new.png', dpi=150)
```
